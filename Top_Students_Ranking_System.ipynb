```python
#To convert the uploaded file to DataFrame(Table)
import pandas as pd
df= pd.read_csv("student_scores.csv")
df.head()
```

```python
import matplotlib.pyplot as plt

#Now this is wher we start the main work, we are to find the mean, median and correlation but the I will add averageto
student_scores_mean= df[['Math','English', 'Science', 'History']].mean()
student_scores_median= df[['Math','English', 'Science', 'History']].median()
student_scores_correlation= df[['Math','English', 'Science', 'History']].corr()

#To display general students, mean medain and correlation
print("Mean of all students score in general:")
print(student_scores_mean)
print("\nMedian of all students score in general:")
print(student_scores_median)
print("\nCorrelation:")
print(student_scores_correlation)

#Now to know each students average

df['mean'] = df[['Math','English', 'Science', 'History']].mean(axis=1)
df['median'] = df[['Math','English', 'Science', 'History']].median(axis=1)

df.head()
 #Now for more visualization, we are plotting df['Name'] against df['mean]
plt.bar(df['Name'], df['mean'])
plt.xlabel('Name of students')
plt.ylabel('Average score')
plt.title('Average score of each student')
plt.show()


```
